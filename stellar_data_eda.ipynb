```python
# split the data
x = df.drop(['class'], axis=1)
y = df['class']
y = y.map({'GALAXY': 0, 'STAR': 1, 'QSO': 2})
# train and test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# scale
scaler = StandardScaler()
scaler.fit(x_train)
x_train = pd.DataFrame(scaler.transform(x_train), columns=x_train.columns)
x_test = pd.DataFrame(scaler.transform(x_test), columns=x_test.columns)
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}

for name, model  in models.items():
    model.fit(x_train, y_train)
    print(f'{name} trained.')
```

```python
for name, model in models.items():
    y_pred = model.predict(x_test)
    print(f'{name} accuracy: {accuracy_score(y_test, y_pred):.4f} | precision: {precision_score(y_test, y_pred, average="micro"):.4f} | recall: {recall_score(y_test, y_pred, average="micro"):.4f} | f1: {f1_score(y_test, y_pred, average="micro"):.4f} | mae: {mean_absolute_error(y_test, y_pred):.4f}')
```
